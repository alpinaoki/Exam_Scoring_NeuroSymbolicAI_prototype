```python
!pip install google-generativeai

import google.generativeai as genai

genai.configure(api_key="YOUR_API_KEY")  # 自分のAPIキーに置き換え

model = genai.GenerativeModel("gemini-1.5-flash")

def extract_structure(question, answer):
    prompt = f"""
あなたは数学の記述答案の構造を抽出するAIです。
問題：
{question}

回答：
{answer}

以下の形式で返してください：
- 主張
- 根拠
- ステップ
- 不足している論理
"""
    res = model.generate_content(prompt)
    return res.text

print(extract_structure(
    "△ABCにおいて、角A=角Bを示せ。",
    "AC=BCだから角Aと角Bが等しい。"
))
```
